**Exemplo - Estatítisca Descritiva**

```python
A = float(input())
B = float(input())
C = float(input())
PA = 2
PB = 3
PC = 5
media = (PA*A + PB*B + PC*C)/(PA + PB + PC)
print(f"Media: {media:.1f}")
```

**Exemplo - Cálculo de variância e desvio padrão**

```python
n1, n2, n3, n4 = input("Insira os números da distribuição: ").split()
n1 = int(n1)
n2 = int(n2)
n3 = int(n3)
n4 = int(n4)
```

```python
# Média
media = (n1 + n2 + n3 + n4)/4
media
```

```python
# Desvios
d1 = n1 - media
d2 = n2 - media
d3 = n3 - media
d4 = n4 - media
d1, d2, d3, d4
```

```python
# Somatória do quadrado dos desvios
soma = d1**2 + d2**2 + d3**2 + d4**2
# Variância da distribuição
variancia = soma / 4
variancia
```

```python
# Desvio Padrão
desvio_padrao = variancia**0.5
desvio_padrao
```

```python
m1, m2, m3, m4 = input("Insira os valores da distribuição: ").split()
m1 = int(m1)
m2 = int(m2)
m3 = int(m3)
m4 = int(m4)
# Média
media1 = (m1 + m2 + m3 + m4) / 4
# Desvios
d10 = m1 - media1
d20 = m2 - media1
d30 = m3 - media1
d40 = m4 - media1
# Variância ----------> Populacional (qdo temos todos os dados observados)
variancia1 = (d10**2 + d20**2 + d30**2 + d40**2) / 4
# Desvio Padrão
desvio1 = variancia1**0.5
print(f"A variância da distribuição {m1}, {m2}, {m3} e {m4} será de: \n{variancia1:.2f}")
print(f"O desvio padrão dessa distribuição será:\n{desvio1:.3f}")
```

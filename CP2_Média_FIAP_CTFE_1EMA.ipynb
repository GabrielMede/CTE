# **Desafio: Cálculo da média anual FIAP**

```python
# Cálculo da média dos checkpoints
# 3 CPs obrigatórios -----> input
# Eliminar a menor nota
# Calcular média das duas maiores notas
```

```python
CP1, CP2, CP3 = input("Insira a nota dos checkpoints: ").split()
CP1 = float(CP1)
CP2 = float(CP2)
CP3 = float(CP3)
if CP1 > CP2 > CP3:
  MCP1 = (CP1 + CP2)/2
elif CP1 > CP2 < CP3:
  MCP1 = (CP1 + CP3)/2
elif CP1 < CP2 < CP3:
  MCP1 = (CP2 + CP3)/2
```

```python
# Cálculo de notas de SPRINTs
# Média das duas notas
SP1, SP2 = input("Insira a nota das Sprints: ").split()
SP1 = float(SP1)
SP2 = float(SP2)
MSP1 = (SP1 + SP2)/2

NF1 = 6 - (0.2*MSP1 + 0.2*MCP1)
print(f"Nota faltante para a média: {NF.:2f}")
```

```python
# Cálculo de notas da Global Solutions
# Inserir a nota
# Caso não haja nota de GS, o aluno deverá inserir sua nora de SUB
GS1 = input("Insira a nota das Global Solutions, caso não tenha digite 0: ").split()
GS1 = float(GS1)
if GS1 == 0:
  input("Insira a nota da substitutiva: ").split()
GS1 = float(GS1)
```

```python
# Cálculo da média semestral 1 (MS1)
# MS1 = 0.6*GS1 + 0.2*MSP1 + 0.2*MCP1
MS1 = (0.6*GS1 + 0.2*MSP1 + 0.2*MCP1)/10
```

```python
# Repetir para o segundo semestre
# MS2
CP4, CP5, CP6 = input("Insira a nota dos checkpoints do segundo semestre: ").split()
CP4 = float(CP4)
CP5 = float(CP5)
CP6 = float(CP6)
if CP4 > CP5 > CP6:
  MCP2 = (CP4 + CP5)/2
elif CP4 > CP5 < CP6:
  MCP2 = (CP4 + CP6)/2
elif CP4 < CP5 < CP6:
  MCP2 = (CP5 + CP6)/2

SP3, SP4 = input("Insira a nota das Sprints do segundo semestre: ").split()
SP3= float(SP3)
SP4 = float(SP4)
MSP2 = (SP3 + SP4)/2

NF2 = 6 - (0.2*MSP2 + 0.2*MCP2)
print(f"Nota faltante para a média: {NF2.:2f}")

GS2 = input("Insira a nota das Global Solutions do segundo semestre, caso não tenha digite 0: ").split()
GS2 = float(GS2)
if GS2 == 0:
  input("Insira a nota da substitutiva: ").split()
GS2 = float(GS2)

MS2 = (0.6*GS2 + 0.2*MSP2 + 0.2*MCP2)/10
```

```python
# Média Final FIAP
MF = 0.4*MS1 + 0.6*MS2
```

```python
# Apresentar se o aluno está aprovado, reprovado ou de exame.
# Aprovado >= 6.0
# Exame >=4.0 e < 6.0
# Reprovado < 4.0
# Se estiver de exame, inserir nota do exame
if MF >= 6.0:
  print("Aprovado")
elif 4.0 <= MF < 6.0:
  print("Exame")
  EX = input("Insira a nota do exame: ")
  EX = float(EX)
  MF2 = (EX + MF)/2
  print(f"Nota Final: {MF2.:2f}")
elif MF < 4.0:
  print("Reprovado")

# Calcule MF2 = (EX + MF)/2
```
